```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Assuming df1 is your DataFrame with 'reviewText' and 'summary' columns
# Concatenate the text from both columns
all_text = ' '.join(af['reviewText'].dropna()) + ' '.join(af['summary'].dropna())

# Tokenization
tokens = word_tokenize(all_text.lower())

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

# Word frequency analysis
word_freq = Counter(filtered_tokens)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Plot word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Review Text and Summary')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data

# Convert 'unixReviewTime' to datetime
af['review_date'] = pd.to_datetime(af['unixReviewTime'], unit='s')

# Group by month and calculate review count or average rating
monthly_data = af.groupby(pd.Grouper(key='review_date', freq='M')).agg({
    'asin': 'count',  # Change 'asin' to the column you want to analyze (e.g., 'overall' for average rating)
})

# Plot time series graph
plt.figure(figsize=(10, 6))
plt.plot(monthly_data.index, monthly_data['asin'], marker='o', linestyle='-')
plt.title('Monthly Review Counts')  # Change title accordingly
plt.xlabel('Date')
plt.ylabel('Review Count')  # Change ylabel accordingly (e.g., 'Average Rating')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
