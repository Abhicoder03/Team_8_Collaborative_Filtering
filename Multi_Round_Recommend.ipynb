```python
import pandas as pd
import gzip
import json

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

af = getDF('Grocery_and_Gourmet_Food_5.json.gz')#C:\Users\srush\Downloads\Luxury_Beauty.json.gz
```

```python
def split_group(group, split_ratio=0.8):
    n_total = len(group)
    n_train = int(n_total * split_ratio)
    idx = group.index
    train_idx = idx[:n_train].tolist()
    test_idx = idx[n_train:].tolist()
    return train_idx, test_idx
```

```python
grouped = rating.groupby('reviewerID')
indices = grouped.apply(split_group)

# concatenate the train and test indices for each group
train_idx = pd.Index([idx for indices in indices for idx in indices[0]])
test_idx = pd.Index([idx for indices in indices for idx in indices[1]])

# create the train and test dataframes
train_df = rating.loc[train_idx]
test_df = rating.loc[test_idx]
```

```python
# Normalize user-item matrix
train_pt_norm = train_pt.subtract(train_pt.mean(axis=1), axis = 'rows').fillna(0)
train_pt_norm.head()
```

```python
from sklearn.metrics.pairwise import pairwise_distances

item_similarity=1-pairwise_distances(train_pt_norm.T,metric='cosine')
print(item_similarity)
item_similarity=item_similarity.argsort()
```

```python
import numpy as np

def input_user(mv=100):
    n = int(input('Enter the number of recommendations you want: '))
    liked_products = []
    liked_product_indices = []  # Store indices of liked products

    print('Products similar to the ones you liked:')
    mv = mv + 1

    rec_items = np.array((item_similarity[[mv], :n]))
    import matplotlib.pyplot as plt
    for mr in np.nditer(rec_items):
        for index, rows in rating.iterrows():
            if index == mr:
                print(f'Index: {index}')
                print(f'ASIN: {rows["asin"]}')
                print(f'Review: {rows["reviewText"]}')
                print()  # Print an empty line for spacing
                liked_products.append(rows['asin'])
                liked_product_indices.append(index)  # Add index to the list

    return liked_products, liked_product_indices  # Return both the list of liked products and their indices


product_list, product_indices = input_user()

while True:
    ans = input("Do you want to update the recommendation list Y/N ?")
    if ans == 'Y':
        x = input("Enter the index of the product you like from above: ")
        x = int(x)  # Convert user input to integer
        if x in product_indices:  # Check if the index is in the list of liked product indices
            product_list.clear()
            product_list, product_indices = input_user(mv=x)  # Pass the index when updating the list
        else:
            print("Index not in the list")
    else:
        break
```
